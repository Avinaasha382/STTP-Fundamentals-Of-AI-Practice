```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris (1).csv")
df
```

```python
df = df[(df["Species"] == "Iris-setosa") | (df["Species"] == "Iris-virginica")][["PetalLengthCm", "SepalWidthCm","Species"]]
df
```

```python
a = np.zeros((3,1))
a
```

## PERCEPTRON ALOGRITHM

```python
i = 0
start = 0
```

```python
while True:
    curr_row = df.iloc[i]
    y = np.array([[curr_row["PetalLengthCm"]],[curr_row["SepalWidthCm"]],[1]])
    print("y",y)
    if curr_row["Species"] == "Iris-virginica":
        y = -1*y
    value = np.matmul(a.transpose(),y)[0]
    print("value",value)
    if value > 0:
        i = (i+1)%100
        if i == start:
            break
    else:
        a = a + 0.01*y
        start = i
        i = (i+1)%100
```
