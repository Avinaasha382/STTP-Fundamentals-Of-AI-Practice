```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv("face_feature_vectors.csv")
df
```

```python
male = df.iloc[0:395,2:] #getting the male training data
male
```

```python
female = df.iloc[400:795,2:]  #getting the female training data
female
```

## COMPUTING THE COVARIANCE MATRICES

```python
male_cov = male.cov()  #computing the covariance matrix
male_cov
```

```python
female_cov = female.cov()  #computing the covariance matrix
female_cov
```

```python
inv_cov_male = np.linalg.inv(male_cov)
inv_cov_male
```

```python
inv_cov_female = np.linalg.inv(female_cov)
inv_cov_female
```

## CASE 3 OF BAYES CLASSIFIER

```python
W1 = -0.5*inv_cov_male
W1
```

```python
W2 = -0.5*inv_cov_female
W2
```

```python
mean1 = male.mean()
mean1
```

```python
mean2 = female.mean()
mean2
```

```python
w1 = np.matmul(inv_cov_male,mean1)
w1
```

```python
w2 = np.matmul(inv_cov_female,mean2)
w2
```

```python
constant_1 = -0.5*(np.matmul(np.matmul(mean1.transpose(),inv_cov_male),mean1))
constant_1
```

```python
constant_2 = -0.5*(np.matmul(np.matmul(mean2.transpose(),inv_cov_female),mean2))
constant_2
```

## TESTING DATA AND BAYES CLASSIFICATION

```python
testing_data = df.iloc[395:400,2:]  #male testing data
testing_data
```

```python
#we substitute the feature vector X in the case 3 of the Bayes decision theory and we get 100% accuracy
for i in testing_data.index:
    curr_row = testing_data.loc[i]
    g1 = np.matmul(np.matmul(curr_row.transpose(),W1),curr_row) + np.matmul(w1.transpose(),curr_row) + constant_1
    g2 = np.matmul(np.matmul(curr_row.transpose(),W2),curr_row) + np.matmul(w2.transpose(),curr_row) + constant_2
    if g1 > g2:
        print("male")
    else:
        print("female")
```
